```python
from cobra import Reaction
reaction = Reaction('rxn00249')
reaction.name = ('malate:NADP+oxidoreductase')
reaction.lower_bound = 0
reaction.upper_bound = 1000
reaction.gene_reaction_rule = '(DESTE_RS08355)'

from cobra import Metabolite
cpd00005_c0 = Metabolite('cpd00005_c0', formula = 'C21H26N7O17P3', name = 'NADPH', compartment = 'c0')
cpd00067_c0 = Metabolite('cpd00067_c0', formula = 'H', name = 'H+', compartment = 'c0')
cpd00003_c0 = Metabolite('cpd00003_c0', formula = 'C21H26N7O14P2', name = 'NAD', compartment = 'c0')
cpd00032_c0 = Metabolite('cpd00032_c0', formula = 'C4H2O5', name = 'Oxaloacetate', compartment = 'c0')
cpd00130_c0 = Metabolite('cpd00130_c0', formula = 'C4H4O5', name = 'L_malate', compartment = 'c0')


reaction.add_metabolites({cpd00005_c0: -1,
                          cpd00067_c0: -1,
                          cpd00032_c0: -1,
                          cpd00003_c0: 1,
                          cpd00130_c0: 1
                          })

reaction.reaction
model.add_reactions([reaction])
```
